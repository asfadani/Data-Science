**SET UP Library yang DIbutuhkan**

```python
import os
from dotenv import load_dotenv # Library untuk membaca file .env
from googleapiclient.discovery import build # library untuk menghubungkan API key
import pandas as pd
```

Mengambil komentar dari video '
NASIB KDMP DESA BENTANGAN KLATEN YANG DULU DIRESMIKAN PRESIDEN & RIBUAN PEJABAT'
https://youtu.be/MO8mDknKLMk?si=_qzqB_pwk1y9Pt37

```python
# Membaca variabel dari file .env
load_dotenv()

# Mwngambil API key  dari video di youtube
api_key = os.getenv('YOUTUBE_API_KEY')
video_id = 'MO8mDknKLMk'

# Membangun koneksi ke API Youtube
youtube = build('youtube', 'v3', developerKey=api_key)

def get_comments(youtube, video_id):
    comments_data = []
    next_page_token = None #Inisialisasi token dengan none untuk tarikan pertama
    # 3. Buat permintaan ke server
    while True :
        request = youtube.commentThreads().list(
            part="snippet", # snippet = berisi informasi inti sebuah komentar
            videoId=video_id,
            maxResults=100, # Jumlah maaksimal komentar yang dapat di ambil
            pageToken=next_page_token # Digunakan agar bisa mengambil lebih dari 100 komentar dengan lanjut ke halaman selanjutntya
        )
        response = request.execute()
    
        # 4. Ekstrak data dari format JSON ke dalam list
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments_data.append({
                'Tanggal' : comment['publishedAt'],
                'Jumlah Like' : comment['likeCount'],
                'Komentar' : comment['textDisplay']
            })
    
    # Mengambil tokrn untuk halaman selanjutnya dari response server
        next_page_token = response.get('nextPageToken')

        if not next_page_token:
            break 

    return comments_data

# Jalankan fungsi
data = get_comments(youtube, video_id)
```

```python
# 5. Rapikan data menggunakan Pandas DataFrame
df = pd.DataFrame(data, columns=['Tanggal', 'Jumlah Like', 'Komentar'])

# Tampilkan 5 baris pertama
print(df.head())

print(df.info())
```

```python
# (Opsional) Simpan ke file JSON
df.to_json('komentar_youtube_KDMP.json', orient='records', indent=4, force_ascii=False)
print(f"Total {len(df)} berhasil disimpan ke komentar_youtube_KDMP.json")
```
