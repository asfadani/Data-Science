**SET UP Library yang DIbutuhkan**

```python
import os
from dotenv import load_dotenv # Library untuk membaca file .env
from googleapiclient.discovery import build # library untuk menghubungkan API key
import pandas as pd
```

Mengambil komentar dari video '
NASIB KDMP DESA BENTANGAN KLATEN YANG DULU DIRESMIKAN PRESIDEN & RIBUAN PEJABAT'
https://youtu.be/MO8mDknKLMk?si=_qzqB_pwk1y9Pt37

```python
# Membaca variabel dari file .env
load_dotenv()

# Mwngambil API key  dari video di youtube
api_key = os.getenv('YOUTUBE_API_KEY')
video_id = 'MO8mDknKLMk'

# Membangun koneksi ke API Youtube
youtube = build('youtube', 'v3', developerKey=api_key)

def get_comments(youtube, video_id):
    comments_data = []
    next_page_token = None #Inisialisasi token dengan none untuk tarikan pertama
    # Buat permintaan ke server
    while True :
        request = youtube.commentThreads().list(
            part="snippet", # snippet = berisi informasi inti sebuah komentar
            videoId=video_id,
            maxResults=100, # Jumlah maaksimal komentar yang dapat di ambil
            pageToken=next_page_token # Digunakan agar bisa mengambil lebih dari 100 komentar dengan lanjut ke halaman selanjutntya
        )
        response = request.execute()
    
        # Ekstrak data dari format JSON ke dalam list
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments_data.append({
                'Tanggal' : comment['publishedAt'],
                'Jumlah Like' : comment['likeCount'],
                'Komentar' : comment['textDisplay']
            })
    
    # Mengambil tokrn untuk halaman selanjutnya dari response server
        next_page_token = response.get('nextPageToken')

        if not next_page_token:
            break 

    return comments_data

# Jalankan fungsi
data = get_comments(youtube, video_id)
```

```python
# Rapikan data menggunakan Pandas DataFrame
df = pd.DataFrame(data, columns=['Tanggal', 'Jumlah Like', 'Komentar'])

# Tampilkan 5 baris pertama
print(df.head())

print(df.info())
```

```python
# Simpan ke file JSON
df.to_json('komentar_youtube_KDMP.json', orient='records', indent=4, force_ascii=False)
print(f"Total {len(df)} berhasil disimpan ke komentar_youtube_KDMP.json")
```

**Membuat Raw Dataset Komentar**

Menggabungkan ke 3 data komentar untuk menjadikan 1 dataset utuh

```python
# Membaca ketiga file

df_video1 = pd.read_json('komentar_youtube_KDMP.json')
df_video2 = pd.read_json('komentar_youtube_MBG_KDMP.json')
df_video3 = pd.read_json('Komentar video 3.json')
```

```python
# Menggabungkan kedua data (Merge/Concat)

df_gabungan = pd.concat([df_video1, df_video2, df_video3], ignore_index=True)
```

```python
df = pd.DataFrame(df_gabungan)

print(df.info(), "\n")
print(df.head())
```

```python
# Memfilter berdasarkan tanggal

# memastikan kolom tanggal menggunakan tipe data DateTime
df_gabungan['Tanggal'] = pd.to_datetime(df_gabungan['Tanggal'], unit='ms')

# Menetukan parameter tanggal
tanggal_mulai = '2025-06-30'
tanggal_selesai = '2026-06-30'

# Menyaring data berdasarkan parameter tanggal yang telah ditentukan
df_final = df_gabungan[(df_gabungan['Tanggal'] >= tanggal_mulai) & (df_gabungan['Tanggal'] <= tanggal_selesai)]

# Membuat record atribut tanggal terbaca dengan format (YYYY-MM-DD)
df_final['Tanggal'] = df_final['Tanggal'].dt.strftime('%Y-%m-%d')

# Menyimpan hasil akhir data
df_final.to_json('raw dataset komentar.json', orient='records', indent=4, force_ascii=False)
print("Proses scraping data komengtar selesai")
```
